```python
import xarray as xr
import glob
```

```python
path = '/home/shom_simurep/public_no_ftp/PROJETS/IberArgo/Data/GIGATL/DEPTHS/'
```

```python
files = glob.glob(path+'*.nc')
files.sort()
files = files[:-2]
print(files[0],'\n...\n',files[-1])
```

```python
def reduceds(ds):
    dsr = ds[['u','v','depth_rho','ptemp','sal']]
    dsr['depth_rho']=-1*dsr['depth_rho']
    return dsr
```

```python
ds = xr.open_mfdataset(files, preprocess=reduceds)
ds
```
